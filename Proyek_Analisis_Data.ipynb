#### Eksplorasi `product_category_name`, `price`, dan `total_penjualan`

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=mergedf, x='product_category_name', y='price', showfliers=False)
plt.xticks(rotation=90)
plt.title('Distribusi Harga per Kategori Produk')
plt.ylabel('Harga')
plt.xlabel('Kategori Produk')
plt.show()
```

```python
category_summary = (
    mergedf.groupby('product_category_name')
    .agg(total_sold=('order_id', 'size'), avg_price=('price', 'mean'))
    .reset_index()
)

plt.figure(figsize=(8, 6))
sns.scatterplot(data=category_summary, x='avg_price', y='total_sold', hue='product_category_name', legend=False)
plt.title('Harga Rata-rata vs Total Penjualan')
plt.xlabel('Harga Rata-rata')
plt.ylabel('Total Penjualan')
plt.show()
```

```python
# Kelompokkan berdasarkan product_id untuk hitung total transaksi
top_products = (
    mergedf.groupby('product_category_name')
    .agg(
        total_sold=('order_id', 'size'),  # Hitung total transaksi per produk
        avg_price=('price', 'mean'),       # Rata-rata harga produk
    )
    .sort_values(by='total_sold', ascending=False)  # Urutkan berdasarkan total penjualan
    .reset_index()
    .nlargest(10, 'total_sold')
)

# Tampilkan 10 produk terlaris
top_products.head(10)
```

```python
# Ambil daftar kategori dari top produk
categories = top_products['product_category_name'].tolist()

# Pilih kategori berdasarkan indeks
index = 0  # Ganti dengan indeks kategori yang diinginkan
category_name = categories[index]

# Konversi kolom tanggal jika belum dilakukan
mergedf['order_purchase_month'] = pd.to_datetime(mergedf['order_purchase_timestamp']).dt.to_period('M')

# Hitung tren penjualan bulanan untuk kategori yang dipilih
trend = mergedf[mergedf['product_category_name'] == category_name] \
    .groupby('order_purchase_month')['order_id'].size()

# Plot tren penjualan
trend.plot(figsize=(10, 4), marker='o', title=f'Tren Penjualan Bulanan: {category_name}')
plt.ylabel('Total Penjualan')
plt.xlabel('Bulan')
plt.show()

# Cetak daftar kategori top produk
print("Daftar Top Produk:", categories)
```

#### Eksplorasi Waktu Transaksi Pelanggan (`customer frequency` dan `recency`)

```python
# Menentukan tanggal transaksi terakhir per pelanggan
last_transaction_date = mergedf.groupby('customer_id')['order_purchase_timestamp'].max().reset_index()

# Tentukan tanggal referensi, misalnya tanggal terakhir dalam dataset
reference_date = mergedf['order_purchase_timestamp'].max()

# Menghitung Recency (hari sejak transaksi terakhir)
last_transaction_date['recency'] = (reference_date - last_transaction_date['order_purchase_timestamp']).dt.days
```

```python
# Scatter plot antara Recency dan jumlah transaksi
customer_transaction_count = mergedf.groupby('customer_id').size().reset_index(name='frequency')
recency_freq_df = last_transaction_date.merge(customer_transaction_count, on='customer_id')
```

```python
plt.figure(figsize=(12, 6))

# Plot dengan warna berbeda untuk boxplot
sns.boxplot(
    x=(recency_freq_df['recency'] < 365), 
    y=recency_freq_df['recency'], 
    palette=["red", "green"]
)

plt.title('Distribusi Recency Berdasarkan Aktivitas Pelanggan')
plt.xlabel('Status Pelanggan')
plt.ylabel('Recency (Hari Sejak Transaksi Terakhir)')

# Label manual langsung di bawah boxplot
plt.xticks([0, 1], ['Tidak Aktif', 'Aktif'])

plt.show()
```

#### Eksplorasi frekuensi `order_id`, jumlah `payment_value` => RFM 

```python
# Menggabungkan Recency dengan Frequency dan Monetary
rfm = pd.merge(last_transaction_date[['customer_id', 'recency']],
               mergedf.groupby('customer_id').agg(
                   frequency=('order_id', 'count'),
                   monetary=('payment_value', 'sum')
               ).reset_index(),
               on='customer_id')
```

```python
# Menentukan kuantil untuk setiap metrik RFM
rfm['recency_score'] = pd.qcut(rfm['recency'], 4, labels=[4, 3, 2, 1])
rfm['frequency_score'] = pd.qcut(rfm['frequency'].rank(method='first'), 4, labels=[1, 2, 3, 4])
rfm['monetary_score'] = pd.qcut(rfm['monetary'], 4, labels=[1, 2, 3, 4])

# Total Skor RFM
rfm['rfm_score'] = rfm['recency_score'].astype(int) + rfm['frequency_score'].astype(int) + rfm['monetary_score'].astype(int)
```

```python
rfm.head()
```

```python
# Segmentasi Berdasarkan Nilai RFM
def segment_customer(row):
    if row['rfm_score'] >= 10:
        return 'Best Customers'
    elif row['rfm_score'] >= 7 and row['rfm_score']<=9:
        return 'Loyal Customers'
    elif row['rfm_score'] >= 4 and row['rfm_score'] <= 6:
        return 'Potential Loyalists'
    elif row['rfm_score'] >= 1 and row['rfm_score'] <= 3:
        return 'At-Risk Customers'
    else:
        return 'Lost Customers'

# Terapkan fungsi segmentasi
rfm['customer_segment'] = rfm.apply(segment_customer, axis=1)


# Hitung jumlah pelanggan di setiap segmen
segment_counts_cust = rfm['customer_segment'].value_counts()
```
